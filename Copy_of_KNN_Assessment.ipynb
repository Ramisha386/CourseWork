```python
import pandas as pd
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/sample_data/weight-height.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Gender = le.fit_transform(df[['Gender']])
```

```python
df.head()
```

```python
df.tail()
```

# Go Ahead

```python
X=df[['Gender','Height']].values
y=df['Weight']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=df[['Gender','Height']]
y=df['Weight']
x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)

train_score = model.score(x_train, y_train)
test_score = model.score(x_test, y_test)

print("Training Score:", train_score)
print("Testing Score:", test_score)

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

# Calculate MSE for training and testing sets
mse_linear_train = mean_squared_error(y_train, y_pred_train)
mse_linear_test = mean_squared_error(y_test, y_pred_test)

print("MSE on training set:", mse_linear_train)
print("MSE on testing set:", mse_linear_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=3)
x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=42)
neigh.fit(x_train,y_train)

train_score = neigh.score(x_train, y_train)
test_score = neigh.score(x_test, y_test)

print("Training Score:", train_score)
print("Testing Score:", test_score)

y_pred_train = neigh.predict(x_train)
y_pred_test = neigh.predict(x_test)

# Calculate MSE for training and testing sets
mse_knn_train = mean_squared_error(y_train, y_pred_train)
mse_knn_test = mean_squared_error(y_test, y_pred_test)

print("MSE on training set:", mse_knn_train)
print("MSE on testing set:", mse_knn_test)
```

```python
print("Linear Regression MSE (Train):", mse_linear_train)
print("Linear Regression MSE (Test):", mse_linear_test)
print("KNN MSE (Train):", mse_knn_train)
print("KNN MSE (Test):", mse_knn_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
neighc = KNeighborsClassifier(n_neighbors=3)
X=df[['Weight','Height']]
y=df['Gender']
x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=42)
neighc.fit(x_train,y_train)
y_pred_train = neighc.predict(x_train)  # Predictions on training data
y_pred_test = neighc.predict(x_test)    # Predictions on testing data

# Calculate accuracy for training and testing sets
accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print("Accuracy on training set:", accuracy_train)
print("Accuracy on testing set:", accuracy_test)
```

```python
import numpy as np
import pandas as pd
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/sample_data/KNNDataset - Sheet1.csv')
df
```

```python
new_column_values = []
for index, row in df.iterrows():
  distance=abs(22-row[0])
  new_column_values.append(distance)
```

```python
new_column_values
```

```python
df['distance'] = new_column_values
```

```python
df
```

```python
df = df.sort_values(by='distance')
df
```

```python
k=0
sum=0
first_three = df.head(3)
for index, row in df.iterrows():
  sum=sum+row[1]
  k=k+1
  if(k==3):
    break
avg=sum/3
print("Calculated Income (manually) when k=3 and age is 22 is {:.2f}".format(avg))
```

```python
from sklearn.linear_model import LinearRegression
X=df['age '].values.reshape(-1, 1)
y=df['income']
# Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict using Linear Regression
X_new = [[22]]
linear_regression_prediction = model.predict(X_new)
print("Linear Regression Prediction for X=22:", linear_regression_prediction[0])
```
